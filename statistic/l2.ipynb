```python
#Ð¾Ð±ÑŠÑÐ²Ð¸Ð¼ Ð¿ÐµÑ€ÐµÐ¼ÐµÐ½Ð½Ñ‹Ðµ Ð´Ð»Ñ Ð¼Ð°Ñ‚ÐµÐ¼Ð°Ñ‚Ð¸Ñ‡ÐµÑÐºÐ¾Ð³Ð¾ Ð¾Ð¶Ð¸Ð´Ð°Ð½Ð¸Ñ Ð¸ Ð´Ð¸ÑÐ¿ÐµÑ€ÑÐ¸Ð¸ Ñ€Ð°ÑÐ¿Ñ€ÐµÐ´ÐµÐ»ÐµÐ½Ð¸Ñ:

#Ð¼Ð°Ñ‚ÐµÐ¼Ñ‚Ð°Ð¸Ñ‡ÐµÑÐºÐ¾Ðµ Ð¾Ð¶Ð¸Ð´Ð°Ð½Ð¸Ðµ
MU = 0.5

#Ð´Ð¸ÑÐ¿ÐµÑ€ÑÐ¸Ñ
VAR = 0.125
```

```python
#Ð½Ð°Ð¿Ð¸ÑˆÐµÐ¼ Ñ„ÑƒÐ½ÐºÑ†Ð¸ÑŽ, Ð²Ñ‹Ñ‡Ð¸ÑÐ»ÑÑŽÑ‰ÑƒÑŽ ÑÑ€ÐµÐ´Ð½ÐµÐµ Ð°Ñ€Ð¸Ñ„Ð¼ÐµÑ‚Ð¸Ñ‡ÐµÑÐºÐ¾Ðµ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ðµ (Ð¼Ð°Ñ‚ÐµÐ¼Ð°Ñ‚Ð¸Ñ‡ÐµÑÐºÐ¾Ðµ Ð¾Ð¶Ð¸Ð´Ð°Ð½Ð¸Ðµ) Ð½Ð°Ð±Ð¾Ñ€Ð° Ñ‡Ð¸ÑÐµÐ»
#sample - Ð¼Ð°ÑÑÐ¸Ð² Ñ‡Ð¸ÑÐµÐ»

def average(sample):
    #Ð¾Ð±ÑŠÑÐ²Ð¸Ð¼ Ð¿ÐµÑ€ÐµÐ¼ÐµÐ½Ð½ÑƒÑŽ Ð´Ð»Ñ ÑÑƒÐ¼Ð¼Ñ‹ Ñ‡Ð¸ÑÐµÐ»
    sum_ = 0
    
    #Ð½Ð°Ñ…Ð¾Ð´Ð¸Ð¼ ÑÑƒÐ¼Ð¼Ñƒ Ñ‡Ð¸ÑÐµÐ» Ð² Ð¼Ð°ÑÑÐ¸Ð²Ðµ
    for x in sample:
        sum_ += x
    
    #Ð´ÐµÐ»Ð¸Ð¼ ÑÑƒÐ¼Ð¼Ñƒ Ñ‡Ð¸ÑÐµÐ» Ð½Ð° ÐºÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾ Ñ‡Ð¸ÑÐµÐ» Ð¸ Ð¿Ð¾Ð»ÑƒÑ‡Ð°ÐµÐ¼ Ñ‚Ñ€ÐµÐ±ÑƒÐµÐ¼Ñ‹Ð¹ Ñ€ÐµÐ·ÑƒÐ»ÑŒÑ‚Ð°Ñ‚
    result = sum_ / float(len(sample))
    
    return result
```

```python
#Ð½Ð°Ð¿Ð¸ÑˆÐµÐ¼ Ñ„ÑƒÐ½ÐºÑ†Ð¸ÑŽ, Ð²Ñ‹Ñ‡Ð¸ÑÐ»ÑÑŽÑ‰ÑƒÑŽ ÑÑ€ÐµÐ´Ð½ÐµÐµ ÐºÐ²Ð°Ð´Ñ€Ð°Ñ‚Ð¸Ñ‡ÐµÑÐºÐ¾Ðµ Ð¾Ñ‚ÐºÐ»Ð¾Ð½ÐµÐ½Ð¸Ðµ Ð½Ð°Ð±Ð¾Ñ€Ð° Ñ‡Ð¸ÑÐµÐ»
#sample - Ð¼Ð°ÑÑÐ¸Ð² Ñ‡Ð¸ÑÐµÐ»

def standard_deviation(sample):
    #Ð½Ð°Ñ…Ð¾Ð´Ð¸Ð¼ Ð¼Ð°Ñ‚ÐµÐ¼Ð°Ñ‚Ð¸Ñ‡ÐµÑÐºÐ¾Ðµ Ð¾Ð¶Ð¸Ð´Ð°Ð½Ð¸Ðµ Ð½Ð°Ð±Ð¾Ñ€Ð° Ñ‡Ð¸ÑÐµÐ»
    variable1 = average(sample)
    
    #ÑÐ³ÐµÐ½ÐµÑ€Ð¸Ñ€ÑƒÐµÐ¼ Ð½Ð¾Ð²Ñ‹Ð¹ Ð½Ð°Ð±Ð¾Ñ€ ÐºÐ²Ð°Ð´Ñ€Ð°Ñ‚Ð¾Ð¼ Ñ‡Ð¸ÑÐµÐ» Ð¸ Ð½Ð°Ð¹Ð´ÐµÐ¼ Ð¸Ñ… Ð¼Ð°Ñ‚ÐµÐ¼Ð°Ñ‚Ð¸Ñ‡ÐµÑÐºÐ¾Ðµ Ð¾Ð¶Ð¸Ð´Ð°Ð½Ð¸Ðµ 
    new_sample = [x**2 for x in sample]
    variable2 = average(new_sample)
    
    #Ð²Ñ‹Ñ‡Ð¸ÑÐ¸Ð¼ ÑÑ€ÐµÐ´Ð½ÐµÐµ ÐºÐ²Ð°Ð´Ñ€Ð°Ñ‚Ð¸Ñ‡ÐµÑÐºÐ¾Ðµ Ð¾Ñ‚ÐºÐ»Ð¾Ð½ÐµÐ½Ð¸Ðµ
    result = math.sqrt(variable2 - variable1 * variable1)
    
    return result
    
```
